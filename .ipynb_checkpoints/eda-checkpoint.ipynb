```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Product Information.csv')
data.head()
```

Clustering analysis for the application quota

```python
columns = ['Kuota Aplikasi (GB)', 'Kuota Videomax (GB)', 'Kuota Maxstream (GB)', 'Kuota Youtube (GB)']
for col in columns:
    data[columns[0]] =  data[columns[0]].fillna(data[col])
data = data.drop(columns=columns[1:])
data['Internet Siang (GB)'] = data['Internet Siang (GB)'].fillna(0)
data['Internet Malam (GB)'] = data['Internet Malam (GB)'].fillna(0)
data['Kuota Utama (GB)'] = data['Kuota Utama (GB)'].fillna(data['Internet Siang (GB)'] + data['Internet Malam (GB)'])
data['Kuota Zona (GB)'] = data['Kuota Zona (GB)'].fillna(0)
data['Kuota Utama (GB)'] = data['Kuota Utama (GB)'] + data['Kuota Zona (GB)']
data = data.drop(columns=['Internet Siang (GB)', 'Internet Malam (GB)', 'Kuota Zona (GB)'])
```

```python
data.to_csv('eda.csv', index=False)
```
